```python
from openai import OpenAI
from pydub import AudioSegment
import os
import io
from dotenv import load_dotenv
load_dotenv("config.env")

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
content = [
    {
        "A": "各位听众朋友大家好，欢迎收听本期科技访谈节目。今天我们邀请到了 SynthMind AI 的技术负责人毓丞先生，来跟我们聊一聊 AI Agent 这个热门话题。毓丞你好！",
        "B": "主持人好，各位听众朋友大家好！很高兴来到节目和大家一起探讨 AI Agent。"
    },
    {
        "A": "最近 AI Agent 可谓是出尽了风头，不少人认为它是通往通用人工智能的关键一步。毓丞，你能先简单介绍一下，到底什么是 AI Agent 吗？",
        "B": "好的。AI Agent 就是能够像人类一样感知环境、做出决策并采取行动的人工智能体。它们可以是软件程序，也可以是具身机器人。与传统的 AI 系统不同的是，AI Agent 更加自主、灵活，能够在复杂多变的环境中独立运作。"
    },
    {
        "A": "听起来很厉害！那 AI Agent 具体有哪些特点呢？",
        "B": "AI Agent 主要有四大特点 : 自主性：AI Agent 能够在没有人类干预的情况下独立运行，自主地做出决策和采取行动。比如 Auto-GPT，你只需要给它设定目标，它就能自己制定计划并执行，最终达成目标。 反应性：AI Agent 能够感知周围环境的变化并做出反应。比如在游戏中，AI Agent 可以根据对手的行动来调整自己的策略。 主动性：AI Agent 不仅被动地响应环境，还能主动地采取行动来改变环境。比如一个学习型 AI Agent，它会主动搜索学习资料来提升自己的技能。 社交能力：AI Agent 能够与其他 AI Agent 或者人类进行沟通和协作。比如在软件开发中，多个 AI Agent 可以通过自然语言交流，共同完成任务。"
    },
    {
        "A": "哇，AI Agent 的能力真的很全面！那它目前有哪些应用场景呢？",
        "B": "AI Agent 的应用场景非常广泛，比如： 基于规则的 Agent： 这是一种比较简单的 AI Agent 类型，它们的决策和行动遵循预先定义的规则。例如，可以用于交通信号灯控制、自动化生产线等场景。 反应式 Agent： 这种 AI Agent 会对环境刺激做出直接反应，而不需要进行复杂的规划或推理。例如，可以用于机器人导航、游戏 AI 等场景。 目标导向的 Agent： 这种 AI Agent 具有明确的目标，会利用规划和决策算法来实现这些目标，例如，可以用于棋类游戏、路径规划等场景。 学习型 Agent： 这种 AI Agent 能够从经验中学习并随着时间的推移改进其性能，例如，可以用于推荐系统、个性化教育等场景。"
    },
    {
        "A": "听起来 AI Agent 的内部机制非常复杂！它究竟是如何工作的呢？",
        "B": "AI Agent 的工作原理可以类比于人类的大脑。它通常包含几个核心模块： ‘大脑’模块： 这个模块是 AI Agent 的核心，它负责高层次的认知和决策，比如理解人类的指令、制定计划、进行推理等等。大型语言模型（LLM）通常是这个模块的核心组件，为 AI Agent 提供强大的语言理解和生成能力。 ‘感知’模块： 这个模块负责接收和处理来自外部环境的信息，比如文本、图像、声音等等。它就像 AI Agent 的眼睛、耳朵和鼻子，帮助 AI Agent 理解周围的环境。 ‘行动’模块： 这个模块负责将决策付诸实践，比如控制机器人的运动、生成文本、使用工具等等。它就像 AI Agent 的手和脚，让 AI Agent 能够在真实世界或虚拟世界中执行各种任务。 ‘记忆’模块： 这个模块负责存储 AI Agent 的经验和知识，供‘大脑’模块进行推理和计划时参考。它就像 AI Agent 的大脑记忆库，帮助 AI Agent 从过去的经验中学习，并在未来做出更好的决策。 ‘计划’模块： 这个模块负责制定实现目标的具体步骤，并将高级计划分解成可执行的行动序列。它就像 AI Agent 的计划员，帮助 AI Agent 高效地完成任务。 这些模块相互协作，使 AI Agent 能够像人类一样处理信息、做出决策并采取行动。"
    },
    {
        "A": "哇，真是太神奇了！看来 AI Agent 未来发展潜力巨大啊！那您认为 AI Agent 未来的发展趋势如何呢？",
        "B": "我认为 AI Agent 的未来发展趋势主要有以下几个方面: 基于 LLM 的 Agent： 未来会有更多 AI Agent 将大型语言模型（LLM）作为其核心组件，赋予它们更强的语言理解和生成能力，使其能够更好地理解人类的指令，并与人类进行更自然、更流畅的交流。 多模态 Agent： 未来的 AI Agent 将能够处理多种模态的信息，比如图像、声音、视频等等，这将使它们能够更全面地感知世界，并完成更复杂的任务。 具身 Agent： 将 AI Agent 的智能与物理世界相结合，让它们能够在真实环境中执行任务，比如控制机器人在现实世界中导航或操作物体，将会是一个重要的发展方向。 Agent 社会： 探索 AI Agent 之间的交互和社会现象，研究如何构建更智能、更协作的 AI 系统，也是未来发展的重要方向。"
    },
    {
        "A": "非常感谢毓丞的精彩分享！相信随着 AI 技术的不断发展，AI Agent 会在未来发挥越来越重要的作用，为我们的生活带来更多便利和惊喜。",
        "B": "谢谢主持人！也感谢各位听众的收听！"
    }
]

def split_text(text, max_length=4096):
    """将文本拆分为多个块，每个块不超过指定的最大长度。"""
    return [text[i:i + max_length] for i in range(0, len(text), max_length)]

def generate_speech_chunk(client, model, input_text, voice, response_format='mp3', speed=1.0):
    """使用 OpenAI 的 TTS API 为单个文本块生成语音。"""
    response = client.audio.speech.create(
        model=model,
        voice=voice,
        input=input_text,
        response_format=response_format,
        speed=speed
    )

    return response.content

def merge_audio_files(audio_files, output_path):
    """合并多个音频文件为一个完整的音频文件。"""
    combined_audio = AudioSegment.empty()
    
    for audio_content in audio_files:
        segment = AudioSegment.from_file(io.BytesIO(audio_content), format="mp3")
        combined_audio += segment

    combined_audio.export(output_path, format="mp3")
    print(f"音频文件已保存为 {output_path}")

def generate_podcast_from_content(client, content, podcast_title, model, voice_a, voice_b):
    """基于对话内容生成完整的播客音频。"""
    # 确保输出目录存在
    output_dir = "/Users/ycyang/code/test/data/audio/"
    os.makedirs(output_dir, exist_ok=True)
    audio_path = os.path.join(output_dir, f"{podcast_title}.mp3")

    audio_files = []

    for i, conversation in enumerate(content):
        # 为每个发言者生成语音
        for speaker, text in conversation.items():
            print(f"生成第 {i+1} 段 {speaker} 的语音")
            try:
                # 使用不同的声音生成不同发言者的语音
                voice = voice_a if speaker == 'A' else voice_b
                audio_content = generate_speech_chunk(client, model, text, voice)
                audio_files.append(audio_content)
            except Exception as e:
                print(f"生成语音失败，第 {i+1} 段 {speaker}：{e}")

    # 合并音频片段
    merge_audio_files(audio_files, audio_path)


# 使用示例
model = 'tts-1'
voice_a = 'nova'  # 主持人的声音
voice_b = 'onyx'   # 受访者的声音

generate_podcast_from_content(client, content, "第四小组科技分享播客", model, voice_a, voice_b)
```
